## Generation of a single structure depiction with polygon coordinates around the chemical structure

```python
smiles = 'CN1C=NC2=C1C(=O)N(C(=O)N2C)C'
# smiles = 'CN1C=NC2=C1C(=O)N(C(=O)N2C)CC(Cc1ccccc1)C(=O)(O)'
dataset_creator = CSD_creator([smiles], load_PLN=True)
```

```python
image, annotation = dataset_creator.generate_structure_with_annotation(smiles)
illustrate_annotations(image, annotation)
```

## Generation of a single structure depiction and a label with polygon coordinates around both elements

```python
image, annotation = dataset_creator.generate_structure_with_annotation(smiles,
                                                                       label=True)
illustrate_annotations(image, annotation)

```
